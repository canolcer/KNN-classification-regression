*1.2.2 Implementing k Nearest Neighbor*

```python
def euclidean_distance(x1, x2):
    return np.sqrt(np.sum((x1 - x2) ** 2))

class KNN:
    def __init__(self, k=3):
        self.k = k

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        predictions = [self._predict(x) for x in X_test]
        return np.array(predictions)

    def _predict(self, x):
        # Calculate distances to all data points in the training set
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]

        # Get indices of k-nearest training data points
        k_neighbors_indices = np.argsort(distances)[:self.k]

        # Get the labels of the k-nearest training data points
        k_neighbor_labels = [self.y_train[i] for i in k_neighbors_indices]

        # Return the most common class label
        most_common = np.bincount(k_neighbor_labels).argmax()
        return most_common

class WeightedKNN(KNN):
    def _predict(self, x):
        # Calculate distances to all data points in the training set
        distances = [euclidean_distance(x, x_train) for x_train in self.X_train]

        # Get indices of k-nearest training data points
        k_neighbors_indices = np.argsort(distances)[:self.k]

        # Get the labels and distances of the k-nearest training data points
        k_neighbor_labels = [self.y_train[i] for i in k_neighbors_indices]
        k_neighbor_distances = [distances[i] for i in k_neighbors_indices]

        # Calculate weighted votes based on distances
        try:
          weighted_votes = np.zeros(np.max(self.y_train) + 1)
        except TypeError:
          weighted_votes = np.zeros(np.max(self.y_train.astype(int)) + 1, dtype=int)

        for label, distance in zip(k_neighbor_labels, k_neighbor_distances):
          try:
            weighted_votes[label] += 1 / (distance ** 2)
          except IndexError:
            pass

        # Return the label with the highest weighted vote
        return np.argmax(weighted_votes)

def calculate_accuracy(true_labels, predicted_labels):
    correct_predictions = sum(true_labels == predicted_labels)
    total_predictions = len(true_labels)
    accuracy = correct_predictions / total_predictions
    return accuracy

def calculate_precision(true_labels, predicted_labels, positive_label):
    true_positive = sum((true_labels == positive_label) & (predicted_labels == positive_label))
    false_positive = sum((true_labels != positive_label) & (predicted_labels == positive_label))
    precision = true_positive / (true_positive + false_positive) if (true_positive + false_positive) != 0 else 0
    return precision

def calculate_recall(true_labels, predicted_labels, positive_label):
    true_positive = sum((true_labels == positive_label) & (predicted_labels == positive_label))
    false_negative = sum((true_labels == positive_label) & (predicted_labels != positive_label))
    recall = true_positive / (true_positive + false_negative) if (true_positive + false_negative) != 0 else 0
    return recall

def k_fold_cross_validation(X, y, k, model, positive_label):
    """
    Perform k-fold cross-validation.

    Parameters:
    - X: Input features
    - y: Target variable
    - k: Number of folds
    - model: Machine learning model to evaluate
    - positive_label: Label for the positive class

    Returns:
    - List of accuracy, precision, and recall scores for each fold
    """

    # Shuffle the data
    data = np.column_stack((X.toarray(), y))
    np.random.shuffle(data)

    # Split data into k folds
    folds = np.array_split(data, k)

    # Initialize lists to store evaluation metrics for each fold
    accuracy_scores = []
    precision_scores = []
    recall_scores = []

    for i in range(k):
        # Use fold i as the test set and the rest as the training set
        test_set = folds[i]
        train_sets = [fold for j, fold in enumerate(folds) if j != i]
        train_set = np.concatenate(train_sets)

        # Split features and target variable
        X_train, y_train = train_set[:, :-1], train_set[:, -1]
        X_test, y_test = test_set[:, :-1], test_set[:, -1]

        # Train the model
        model.fit(X_train, y_train)

        # Predict labels on the test set
        predicted_labels = model.predict(X_test)

        # Calculate evaluation metrics
        accuracy = calculate_accuracy(y_test, predicted_labels)
        precision = calculate_precision(y_test, predicted_labels, positive_label)
        recall = calculate_recall(y_test, predicted_labels, positive_label)

        # Append scores to the respective lists
        accuracy_scores.append(accuracy)
        precision_scores.append(precision)
        recall_scores.append(recall)

    return accuracy_scores, precision_scores, recall_scores
```

```python
# Function to find the top 3 words and their frequencies for a given category
def top_words_and_frequencies(category_data):
    # Create a CountVectorizer with English stop words
    vectorizer = CountVectorizer(stop_words=list(ENGLISH_STOP_WORDS))

    # Fit and transform the category data
    X = vectorizer.fit_transform([category_data])

    # Get feature names and word frequencies
    feature_names = vectorizer.get_feature_names_out()
    word_freq = X.toarray()[0]

    # Create a dictionary of word frequencies
    word_freq_dict = dict(zip(feature_names, word_freq))

    # Remove stop words from the dictionary
    word_freq_dict = {word: freq for word, freq in word_freq_dict.items() if word not in ENGLISH_STOP_WORDS}

    # Sort the dictionary by frequency in descending order
    sorted_word_freq = sorted(word_freq_dict.items(), key=lambda x: x[1], reverse=True)

    # Get the top 3 words and their frequencies
    top_words = [word for word, freq in sorted_word_freq[:3]]
    top_word_frequencies = [freq for word, freq in sorted_word_freq[:3]]

    return top_words, top_word_frequencies

# Create subplots for each category
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 10))
fig.subplots_adjust(hspace=0.5)

# Find and plot the top 3 words for each category
for i, category in enumerate(numeric_map.keys()):
    category_data = ' '.join(numeric_data_frame[numeric_data_frame['Category'] == numeric_map[category]]['Text'])
    top_words, top_word_frequencies = top_words_and_frequencies(category_data)

    # Create a bar plot for the current category
    axes[i // 3, i % 3].bar(top_words, top_word_frequencies)
    axes[i // 3, i % 3].set_title(f'Top 3 Words in Category {category}')
    axes[i // 3, i % 3].set_xlabel('Words')
    axes[i // 3, i % 3].set_ylabel('Frequency')

plt.show()
```

```python
def normalize(data):
  df = data.copy()
  cols = df.columns
  for col in cols:

    maxi = max(df[col])
    mini = min(df[col])

    res = (df[col] - mini) / (maxi - mini)

    df[col] = res

  return df

```

```python
def MEA(pred,actual):
  n = len(pred)
  sum = 0
  for pre in range(len(pred)):
    d = abs(pred[pre] - actual[pre])
    sum += d

  return sum / n
```

```python
X_train_nonf = X_train.copy().to_numpy()
X_train_fn = normalize(X_train).to_numpy()

y_train_ = y_train.copy().to_numpy()
y_test_ = y_test.copy().to_numpy()
X_test_nonf = X_test.copy().to_numpy()
x_test_fn = normalize(X_test).to_numpy()

```

```python
def kfold_indices(data,test, k):
    fold_size = len(data) // k
    indices = np.arange(len(data))
    folds = []
    for i in range(k):
        test_indices = indices[i * fold_size: (i + 1) * fold_size]
        train_indices = np.concatenate([indices[:i * fold_size], indices[(i + 1) * fold_size:]])
        folds.append((train_indices, test_indices))
    return folds
```

```python
k_values = [1,3,5,7,9]

fold5k_nonf = kfold_indices(X_train_nonf,X_test_nonf,5)

fold5k_fn = kfold_indices(X_train_fn,x_test_fn,5)


print("non normalized")
for k in k_values:
  model_knn = KNN(k)
  model_knn.fit(X_train_nonf,y_train)
  model_wwknn = WeightedKNN(k).fit(X_train_nonf,y_train)

  # Initialize a list to store the evaluation scores
  scores = []

  # Iterate through each fold
  for train_indices, test_indices in fold5k_nonf:
      X_train_kf, y_train_kf = X_train_nonf[train_indices], y_train_[train_indices]
      X_test_kf, y_test_kf = X_test_nonf[test_indices], y_test_[test_indices]

      # Train the model on the training data
      model_knn.fit(X_train_kf, y_train_kf)

      # Make predictions on the test data
      y_pred = model_knn.predict(X_test_kf)

      fold_score = MEA(y_test_kf, y_pred)

      # Append the fold score to the list of scores
      scores.append(fold_score)
      break


  mean_err = np.mean(scores)
  print("For k={}".format(k))
  print("5-Fold Cross-Validation Scores:", scores)
  print("Mean Error:", mean_err)
```
